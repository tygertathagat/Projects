```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots()
dataset[dataset['RainTomorrow'] == 'No'].plot.scatter(x='Humidity3pm', y='Pressure3pm', c='purple', alpha=0.25, ax=ax)
dataset[dataset['RainTomorrow'] == 'Yes'].plot.scatter(x='Humidity3pm', y='Pressure3pm', c='orange', alpha=0.25, ax=ax)
```

```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import accuracy_score
```

```python
dataset_clean = dataset.dropna()
```

```python
len(dataset),len(dataset_clean)
```

```python
X = dataset_clean[['Humidity3pm', 'Pressure3pm']]
y= dataset_clean['RainTomorrow']

y
```

```python
y = np.array([0 if value=='No' else 1 for value in y])
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state= 33)
```

```python
neighbour = KNeighborsClassifier()
neighbour.fit(X_train,y_train)
y_prediction = neighbour.predict(X_test)
accuracy_score(y_test, y_prediction)
```

```python
scores = []
for k in range(1,100):
    neighbour = KNeighborsClassifier(n_neighbors=k)
    neighbour.fit(X_train,y_train)
    y_prediction = neighbour.predict(X_test)
    score = accuracy_score(y_test, y_prediction)
    scores.append(score)
    
```

```python
fig, ax = plt.subplots()
ax.plot(scores)
```
